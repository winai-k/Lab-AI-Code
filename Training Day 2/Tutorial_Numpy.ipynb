## The matrix

$$A = \begin{bmatrix}
a_{11} & a_{12} & a_{13} & \cdots & a_{1N}\\
a_{21} & a_{22} & a_{23} & \cdots & a_{2N}\\
\vdots & \vdots & \vdots & \ddots & \vdots\\
a_{M1} & a_{M2} & a_{M3} & \cdots & a_{MN}\\
\end{bmatrix}$$

## The vector

$$V = \begin{bmatrix}
v_{11} &
v_{12} &
v_{13} &
\cdots &
v_{1M}
\end{bmatrix}$$

````python
np.random.seed(7) # set seed random

M = 2 # กำหนดจำนวน "แถว"
N = 2 # กำหนดจำนวน "หลัก"

A = np.random.randn(M, N) # สร้างเมทริกซ์ "A"
V = np.random.randn(1, M) # สร้างเวคเตอร์ "V"
````

## Inverse matrix

$$AA^{-1} = I$$

เมื่อ $I$ เป็น Identity matrix โดยที่ $AI = A$

code:
```python
A_inv = np.lin.inv(A)
```

Note: A ต้องเป็น square matrix เท่านั้น

````python
A = np.random.rand(5, 5) # สร้าง square matrix หรือ เมทริกซ์ที่มีขนาด แถว และ หลัก เท่ากัน

A_inv = np.linalg.inv(A) # หา Inverse A
````

````python
# ลองหา I กัน
I = A @ A_inv
````

````python
np.round(I)
````

````python
np.round(I) @ A
````

````python
A
````

````python
# หาอินเวิร์สของ A_T @ A
````

````python
A_new = A_T @ A
A_new_i = np.linalg.inv(A_new)
````
