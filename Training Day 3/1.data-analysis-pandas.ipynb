# Missing Data

```python
df=pd.DataFrame({'A':[1,2,None],
                 'B':[5,None,None],
                 'C':[1,2,3]})
```

```python
df
```

```python
df.dropna() #drop แถว
```

```python
df.dropna(axis=1) #drop หลัก
```

```python
df.dropna(thresh=2, axis=1)
```

```python
df.fillna(value='Fill')
```

```python
df['A'].mean()
```

```python
df['A']=df['A'].fillna(value=df['A'].mean()) #เติมค่าเฉลี่ยนใน na
```
